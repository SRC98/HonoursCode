# SVM Model

### Importing necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import keras
from keras.utils import np_utils
from skimage.feature import hog
from skimage.color import rgb2grey
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix , classification_report , accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import roc_curve, auc
import cv2
from PIL import Image
import os
```

```python
infected = os.listdir("input/cell_images/Parasitized/")
uninfected = os.listdir("input/cell_images/Uninfected/")
```

```python
data = []
labels = []

for i in infected:
    try:
    
        image = cv2.imread("input/cell_images/Parasitized/"+i)
        image_array = Image.fromarray(image , "RGB")
        resize_img = image_array.resize((50 , 50))
        data.append(np.array(resize_img))
        labels.append(1)
        
    except AttributeError:
        print("")
        
for u in uninfected:
    try:
        
        image = cv2.imread("input/cell_images/Uninfected/"+u)
        image_array = Image.fromarray(image , "RGB")
        resize_img = image_array.resize((50 , 50))
        data.append(np.array(resize_img))
        labels.append(0)
        
    except AttributeError:
        print("")
```

```python
cells = np.array(data)
labels = np.array(labels)

np.save("Cells-SVM", cells)
np.save("Labels-SVM", labels)
```

```python
print("Cells : {} | Labels : {}".format(cells.shape, labels.shape))
```

```python
# flat_cells = cells.flatten().reshape(96453, 7500) previous image amount
flat_cells = cells.flatten().reshape(27558, 7500)
print (flat_cells.shape)
```

```python
n = np.arange(flat_cells.shape[0])
np.random.shuffle(n)
flat_cells = flat_cells[n]
labels = labels[n]

flat_cells = flat_cells.astype(np.float32)
labels = labels.astype(np.int32)
flat_cells = flat_cells/255
```

```python
ss = StandardScaler()
flat_cells_stand = ss.fit_transform(flat_cells)

pca = PCA(n_components = 500)
flat_cells_pca = pca.fit_transform(flat_cells_stand)

print("PCA cells shape is: ", flat_cells_pca.shape)
```

```python
train_x, test_X, train_y, test_Y = train_test_split(flat_cells_pca, labels, test_size = 0.2, random_state = 115)
```

```python
print("train data shape {}, test data shape {}".format(train_x.shape, test_X.shape))
```

```python
print("train data shape {}, test data shape {}".format(train_y.shape, test_Y.shape))
```

```python
svm_malaria = SVC (max_iter = 20000, random_state=115)

svm_malaria.fit(train_x, train_y)
```

```python
pred_y = svm_malaria.predict(test_X)

accuracy = accuracy_score(test_Y, pred_y)
print('Model accuracy is: ', accuracy)
```

```python
from sklearn import metrics
confusion = metrics.confusion_matrix(test_Y, pred_y)
print(confusion)
TP = confusion[1, 1]
TN = confusion[0, 0]
FP = confusion[0, 1]
FN = confusion[1, 0]
```

```python
sensitivity = TP / float(FN + TP)

print(sensitivity)
```

```python
specificity = TN / (TN + FP)

print(specificity)
```

```python
svm_malaria.decision_function(test_X)[0:10]
```

```python
pred_y_prob = svm_malaria.decision_function(test_X)[:]
```

```python
from sklearn import metrics

fpr, tpr, thresholds = metrics.roc_curve(test_Y, pred_y_prob)

plt.plot(fpr, tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.rcParams['font.size'] = 12
plt.title('ROC curve for malaria classifier')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.grid(True)
```

```python
print(metrics.roc_auc_score(test_Y, pred_y_prob))
```

```python
def evaluate_threshold(threshold):
    print('Sensitivity:', tpr[thresholds > threshold][-1])
    print('Specificity:', 1 - fpr[thresholds > threshold][-1])
```

```python
evaluate_threshold(0.1)
```
